```python
# RENTABILIDAD ANUALIZADA

df_2019 = df[(df['year'] == 2019) & (df['afp'] == 'Cuprum') &(df['fondo'] == 'A')]

def compound_compute(monthly_returns: pd.Series) -> float | None:
    total_return = 1
    for monthly_return in monthly_returns:
        total_return = total_return * (1 + (monthly_return / 100))
    total_return = (total_return - 1) * 100
    return total_return

anual_return = df.groupby(['afp','fondo','year']).agg(anual_return = ('rent_real',compound_compute)).reset_index()

anual_return.tail(20)


anual_return[anual_return['afp'] == 'Capital']



```
